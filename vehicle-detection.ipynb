```python
import yaml
import tempfile
import os
from ultralytics import YOLO

# Define dataset paths and class names using forward slashes
data_dict = {
    'train': '/kaggle/input/kitti-vehicle-dataset/dataset/train',
    'val': '/kaggle/input/kitti-vehicle-dataset/dataset/val',
    'nc': 9,
    'names': ['Cyclist', 'Truck', 'Car', 'DontCare', 'Misc', 'Pedestrian', 'Tram', 'Van', 'Person_sitting']
}

# Create a temporary YAML file
with tempfile.NamedTemporaryFile(delete=False, suffix='.yaml', mode='w') as temp_yaml_file:
    yaml.dump(data_dict, temp_yaml_file)
    temp_yaml_path = temp_yaml_file.name  # Get the path of the temporary file

# Initialize the model
model = YOLO('yolov8n.pt')
```

```python
# Train the model using the temporary YAML file
results = model.train(data=temp_yaml_path, epochs=200, batch=16, imgsz=640)

# Output results
print('Results are : ', results)

# Cleanup: Remove the temporary YAML file
os.remove(temp_yaml_path)
```
